```python
columns=['name','age','gender','job']
user1=pd.DataFrame([['sahithi',20,'F','student'],
                    ['ramu',21,'M','student']],columns=columns)
```

```python
user2=pd.DataFrame([['Pathi',20,'M','student'],
                    ['Durga',21,'F','student']],columns=columns)
user3=pd.DataFrame([['Kishore',20,'M','scientist'],
                    ['krishna',21,'M','manager']],columns=columns)
```

```python
user4=pd.DataFrame(dict(name=['Nani','Depithi'],
                        age=[33,44],gender=['M','F'],
                        job=['engineer','scientist']))
print(user4)
```

```python
user5=pd.DataFrame(dict(name=['sahithi','Pathi','Durga', 'Krishna','Nani','Depithi'],
                        height=[165,180,171,160,184,165]))
print(user5)
```

```python
users=pd.concat([user1,user2,user3,user4])
users
```

```python
users=pd.merge(users,user5,on='name',how='outer')
print(users)
```

```python
'''Consider a DataFrame with columns like Name, Math, and Science. Using melt(),
 you can transform it so that Name remains an identifier, and Math and Science
  are unpivoted into a new Subject column and a Score column.

  Pivot-->Rows to columns when convert the rows into columns one column into multiple cloumns.we used Present information in a new perspective form
  UnPivot -->Collumns to rows eaiser to anlaysis of the dataframe .we used in the easier to perform on a row-by-row basis
'''

data = {'Name': ['Alice', 'Bob'],
        'Math': [90, 85],
        'Science': [88, 92]}
df = pd.DataFrame(data)
print(df)
# Melt the DataFrame
melted_df = pd.melt(df, id_vars=['Name'], var_name='Subject', value_name='Score')

print('melted df',melted_df)
```

```python
'''Use pandas.pivot() when your data strictly adheres to a one-to-one mapping
between the index and columns values, and you don't need to perform any aggregation.'''
#Need to aggregate data
print(melted_df.pivot(index='Name',columns='Subject',values='Score'))
```
