```python
'''
Start by creating a simple agent that can answer questions and call tools. The agent in this example uses the chosen language model, 
a basic weather function as a tool, and a simple prompt to guide its behavior'''

from langchain.agents import create_agent

def get_wheather(location:str) -> str:
    """get weather for a given location"""
    return f"The weather in {location} is sunny."
agent=create_agent(
    model="google_genai:gemini-2.5-flash-lite",
    tools=[get_wheather],
    system_prompt="You are a helpful assistant that can answer questions and provide weather information."
)
result=agent.invoke({"messages":[{"role":"user","content":"What is the weather in New York?"}]})
print(result)
```
